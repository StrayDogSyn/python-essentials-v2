### Step 2: Your Task â€” Improve the Exception Handling

**ğŸ¯ Task:** Improve the code to:
1. Use a **specific exception type** (`ValueError`)
2. Replace `None` with a **default value of `0`**
3. **Count** how many conversions failed
4. **Log** which inputs failed

**ğŸ’¡ Hint:** You can store failed inputs in a separate list!

**ğŸ† Bonus:** Create a function that returns a dictionary with `{'results': [...], 'failed': [...], 'success_rate': X%}`

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ¨ YOUR SOLUTION: Improved Exception Handling
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

inputs = ['10', 'x', '25', '3.14', '', '-5', 'hello']

# Track results and failures
results = []
failed_inputs = []
failed_count = 0

for s in inputs:
    try:
        results.append(int(s))
    except ValueError:                   # âœ… Specific exception type
        results.append(0)                # âœ… Default value instead of None
        failed_inputs.append(s)          # âœ… Track which ones failed
        failed_count += 1

# Calculate success rate
success_rate = ((len(inputs) - failed_count) / len(inputs)) * 100

print("ğŸ“¥ Inputs:", inputs)
print("ğŸ“¤ Results:", results)
print(f"âŒ Failed inputs: {failed_inputs}")
print(f"ğŸ“Š Failed conversions: {failed_count}")
print(f"âœ… Success rate: {success_rate:.1f}%")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ† BONUS: Function that returns a complete report
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def safe_convert_to_int(inputs: list, default: int = 0) -> dict:
    """
    Safely convert a list of strings to integers with full error reporting.
    
    Args:
        inputs: List of strings to convert
        default: Value to use when conversion fails
        
    Returns:
        Dictionary with results, failed items, and success rate
    """
    results = []
    failed = []
    
    for item in inputs:
        try:
            results.append(int(item))
        except ValueError:
            results.append(default)
            failed.append(item)
    
    total = len(inputs)
    success_count = total - len(failed)
    
    return {
        'results': results,
        'failed': failed,
        'success_count': success_count,
        'fail_count': len(failed),
        'success_rate': f"{(success_count / total) * 100:.1f}%" if total > 0 else "N/A"
    }

# Test the function
test_inputs = ['10', 'x', '25', '3.14', '', '-5', 'hello']
report = safe_convert_to_int(test_inputs)

print("ğŸ† BONUS SOLUTION - Complete Report:")
print("=" * 60)
for key, value in report.items():
    print(f"   {key}: {value}")
````

### Step 1: Understanding Generators

A **generator** produces values one at a time using `yield`, instead of creating all values in memory at once.

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“– EXAMPLE: Generator that yields words longer than n characters
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def words_longer_than(words: list, n: int):
    """
    Generator that yields words longer than n characters.
    
    Args:
        words: List of words to filter
        n: Minimum length (exclusive)
        
    Yields:
        Words with length > n
    """
    for word in words:
        if len(word) > n:
            yield word                   # <- YIELD: Produces one value at a time

# Create the generator
sample_words = ['this', 'session', 'is', 'about', 'generators', 'and', 'iteration']
gen = words_longer_than(sample_words, 4)

print("ğŸ“Š Sample words:", sample_words)
print(f"ğŸ¯ Generator object: {gen}")
print(f"ğŸ“¤ Words longer than 4 chars: {list(gen)}")
````

### Step 2: Your Task â€” Create a Reusable Generator Module

**ğŸ¯ Task:** 
1. Define the generator in the cell below
2. In the following cell, import it using `from __main__ import`
3. Test it with different inputs

**ğŸ’¡ Hint:** In Jupyter notebooks, `__main__` refers to the notebook's namespace.

**ğŸ† Bonus:** Add a parameter to optionally transform the words (e.g., uppercase).

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“¦ MODULE CELL: Define the generator here (simulates a separate .py file)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

from typing import Generator, Callable, Optional

def words_longer_than(
    words: list, 
    n: int, 
    transform: Optional[Callable[[str], str]] = None
) -> Generator[str, None, None]:
    """
    Generator that yields words longer than n characters.
    
    Args:
        words: List of words to filter
        n: Minimum length (exclusive)
        transform: Optional function to transform each word (e.g., str.upper)
        
    Yields:
        Words with length > n, optionally transformed
        
    Example:
        >>> list(words_longer_than(['hello', 'hi'], 3))
        ['hello']
        >>> list(words_longer_than(['hello', 'hi'], 3, str.upper))
        ['HELLO']
    """
    for word in words:
        if len(word) > n:
            if transform:
                yield transform(word)
            else:
                yield word

print("âœ… Module loaded! Generator 'words_longer_than' is now available.")
````

# ğŸš€ Code Challenge: Log File Parser

---

## ğŸ¯ Challenge Goal

Create a generator function `parse_log(lines)` that processes log entries and yields structured data.

### Requirements:

| Requirement | Description |
|-------------|-------------|
| **Input** | List of log lines in format `LEVEL:MESSAGE` |
| **Output** | Generator yielding dictionaries `{'level': ..., 'message': ...}` |
| **Error Handling** | Skip malformed lines (no colon), but track them |
| **Bonus** | Filter by log level (e.g., only ERROR entries) |

### Example:

```python
logs = ['INFO:User logged in', 'ERROR:Database connection failed', 'Bad line']
for entry in parse_log(logs):
    print(entry)
# {'level': 'INFO', 'message': 'User logged in'}

```

---

### ğŸ‘¥ Team Approach:

1. **ğŸš— Driver:** Writes the initial function structure
2. **ğŸ§­ Navigator:** Guides the logic and error handling
3. **ğŸ§ª Tester:** Creates test cases including edge cases

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸš€ CODE CHALLENGE: Log File Parser
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# YOUR TEAM'S SOLUTION GOES HERE!

from typing import Generator, Optional

def parse_log(
    lines: list, 
    level_filter: Optional[str] = None
) -> Generator[dict, None, None]:
    """
    Parse log entries and yield structured dictionaries.
    
    Args:
        lines: List of log lines in format 'LEVEL:MESSAGE'
        level_filter: Optional - only yield entries matching this level
        
    Yields:
        Dictionary with 'level' and 'message' keys
        
    Skips:
        Malformed lines (missing colon separator)
    """
    skipped = []
    
    for line in lines:
        # Check if line has the expected format
        if ':' not in line:
            skipped.append(line)
            continue
        
        # Split on first colon only (message might contain colons)
        parts = line.split(':', 1)
        level = parts[0].strip()
        message = parts[1].strip() if len(parts) > 1 else ''
        
        # Apply level filter if specified
        if level_filter and level.upper() != level_filter.upper():
            continue
        
        yield {
            'level': level,
            'message': message
        }
    
    # Report skipped lines at the end (for debugging)
    if skipped:
        print(f"âš ï¸ Skipped {len(skipped)} malformed lines: {skipped}")


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ§ª TEST YOUR SOLUTION
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
sample_logs = [
    'INFO:User logged in',
    'ERROR:Database connection failed',
    'Bad line without separator',
    'WARNING:Low disk space',
    'DEBUG:Processing request: id=123',
    '',
    'ERROR:Permission denied for user admin'
]

print("ğŸ§ª TESTING LOG PARSER")
print("=" * 60)
print("ğŸ“¥ Sample logs:", sample_logs[:3], "...")
print("-" * 60)

print("\nğŸ“¤ All valid entries:")
for entry in parse_log(sample_logs):
    print(f"   [{entry['level']}] {entry['message']}")

print("\nğŸ“¤ Only ERROR entries:")
for entry in parse_log(sample_logs, level_filter='ERROR'):
    print(f"   [{entry['level']}] {entry['message']}")
````
