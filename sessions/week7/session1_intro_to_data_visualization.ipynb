````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ¨ PYTHONIC: Professional line chart with styling
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("âœ¨ PYTHONIC: Professional line chart")
print("-" * 40)

fig, ax = plt.subplots(figsize=(12, 6))

# Plot line with markers
ax.plot(months, monthly_sales, 
        marker='o',           # Add circle markers
        markersize=8,
        linewidth=2,
        color='#3498db',      # Nice blue
        markerfacecolor='white',
        markeredgecolor='#3498db',
        markeredgewidth=2)

# Add a trend line
z = np.polyfit(range(len(monthly_sales)), monthly_sales, 1)
p = np.poly1d(z)
ax.plot(months, p(range(len(monthly_sales))), 
        '--', color='#e74c3c', alpha=0.7, label='Trend')

# Highlight max and min
max_idx = monthly_sales.index(max(monthly_sales))
min_idx = monthly_sales.index(min(monthly_sales))
ax.annotate(f'Peak: ${max(monthly_sales):,}', 
            xy=(max_idx, max(monthly_sales)),
            xytext=(max_idx-1, max(monthly_sales)+3000),
            fontsize=10,
            arrowprops=dict(arrowstyle='->', color='green'))

# Labels and title
ax.set_title('2024 Monthly Sales Performance', fontsize=14, fontweight='bold')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Sales (USD)', fontsize=12)
ax.legend()

# Grid
ax.grid(True, alpha=0.3)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

plt.tight_layout()
plt.show()
print("=" * 50)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸŒ± NOVICE: Basic scatter plot with labels
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ… Better: Adding context with titles and labels

import matplotlib.pyplot as plt

# Study hours vs exam scores
study_hours = [1, 2, 3, 4, 5, 6, 7]
exam_scores = [50, 55, 60, 65, 72, 80, 88]

# Create scatter plot
plt.scatter(study_hours, exam_scores)

# Add labels
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Scores")

plt.show()

print("\nâœ… Now we understand what the data represents!")
print("ğŸ’¡ The scatter shows a positive correlation: more study = higher scores")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”§ INTERMEDIATE: Enhanced scatter with styling and annotations
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ… Better: Colors, sizes, and grid for readability

import matplotlib.pyplot as plt

# Coffee consumption vs productivity
coffee_cups = [0, 1, 2, 3, 4, 5, 6]
productivity = [40, 55, 70, 80, 75, 60, 45]  # Note: too much coffee decreases productivity!

# Create figure with size
fig, ax = plt.subplots(figsize=(8, 5))

# Scatter with custom styling
scatter = ax.scatter(
    coffee_cups, 
    productivity,
    c='brown',           # Coffee color!
    s=100,               # Point size
    alpha=0.7,           # Transparency
    edgecolors='darkbrown',
    linewidth=2
)

# Labels and title
ax.set_xlabel("Coffee Cups per Day", fontsize=12)
ax.set_ylabel("Productivity Score", fontsize=12)
ax.set_title("â˜• Coffee Consumption vs Productivity", fontsize=14, fontweight='bold')

# Add grid
ax.grid(True, linestyle='--', alpha=0.5)

# Annotate the peak
ax.annotate(
    'Optimal!',
    xy=(3, 80),
    xytext=(4.5, 85),
    fontsize=10,
    arrowprops=dict(arrowstyle='->', color='green')
)

plt.tight_layout()
plt.show()

print("\nğŸ” Insights:")
print("  â€¢ Peak productivity at 3 cups")
print("  â€¢ Law of diminishing returns after that!")
print("  â€¢ This shows a quadratic (curved) relationship")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ¨ PYTHONIC: Professional scatter with multiple series and trend line
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ… Best: Multiple categories, numpy trend line, professional styling

import matplotlib.pyplot as plt
import numpy as np

# Sample data: Experience vs Salary by education level
np.random.seed(42)  # For reproducibility

# Generate sample data
n_points = 15

# High School graduates
hs_experience = np.random.uniform(0, 20, n_points)
hs_salary = 30000 + hs_experience * 1500 + np.random.normal(0, 3000, n_points)

# Bachelor's degree
bs_experience = np.random.uniform(0, 20, n_points)
bs_salary = 45000 + bs_experience * 2500 + np.random.normal(0, 4000, n_points)

# Master's degree
ms_experience = np.random.uniform(0, 20, n_points)
ms_salary = 60000 + ms_experience * 3500 + np.random.normal(0, 5000, n_points)

# Create professional figure
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each education level
ax.scatter(hs_experience, hs_salary, c='#3498db', s=80, alpha=0.6, 
           label='High School', marker='o', edgecolors='white')
ax.scatter(bs_experience, bs_salary, c='#2ecc71', s=80, alpha=0.6, 
           label="Bachelor's", marker='s', edgecolors='white')
ax.scatter(ms_experience, ms_salary, c='#9b59b6', s=80, alpha=0.6, 
           label="Master's", marker='^', edgecolors='white')

# Add trend lines using numpy polyfit
for exp, sal, color in [(hs_experience, hs_salary, '#3498db'),
                        (bs_experience, bs_salary, '#2ecc71'),
                        (ms_experience, ms_salary, '#9b59b6')]:
    z = np.polyfit(exp, sal, 1)  # Linear fit
    p = np.poly1d(z)
    x_line = np.linspace(0, 20, 100)
    ax.plot(x_line, p(x_line), color=color, linestyle='--', alpha=0.5, linewidth=2)

# Professional styling
ax.set_xlabel("Years of Experience", fontsize=12, fontweight='bold')
ax.set_ylabel("Annual Salary ($)", fontsize=12, fontweight='bold')
ax.set_title("ğŸ’¼ Experience vs Salary by Education Level", fontsize=14, fontweight='bold')

# Format y-axis as currency
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

# Legend and grid
ax.legend(loc='upper left', framealpha=0.9)
ax.grid(True, linestyle='--', alpha=0.3)

# Set axis limits
ax.set_xlim(-1, 21)
ax.set_ylim(20000, 150000)

plt.tight_layout()
plt.show()

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š Summary Statistics
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
print("\nğŸ“ˆ Scatter Plot Features Used:")
print("  âœ… Multiple data series with different colors")
print("  âœ… Different markers for each category (o, s, ^)")
print("  âœ… Trend lines using numpy.polyfit()")
print("  âœ… Currency formatting on y-axis")
print("  âœ… Professional legend placement")

print("\nğŸ”¬ Key Insights from the Data:")
print("  â€¢ Higher education = higher starting salary")
print("  â€¢ All groups show positive correlation with experience")
print("  â€¢ Master's degree has steepest salary growth")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸŒ± NOVICE: Simple pie chart with percentages
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ… Better: Fewer categories, percentages shown

import matplotlib.pyplot as plt

# Time spent learning programming (per week)
activities = ['Coding Practice', 'Reading Docs', 'Watching Videos', 'Projects']
hours = [10, 3, 5, 7]

# Create pie with percentages
plt.pie(hours, labels=activities, autopct='%1.0f%%')
plt.title("Weekly Programming Learning Time")
plt.show()

print("\nâœ… Improvements:")
print("  â€¢ Only 4 categories (easy to distinguish)")
print("  â€¢ Percentages shown with autopct='%1.0f%%'")
print("  â€¢ Clear title explaining the data")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”§ INTERMEDIATE: Styled pie chart with explosion and shadow
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ… Better: Visual emphasis, colors, and styling

import matplotlib.pyplot as plt

# Browser market share
browsers = ['Chrome', 'Safari', 'Firefox', 'Edge', 'Other']
market_share = [65, 18, 8, 5, 4]
colors = ['#4285F4', '#FF9500', '#FF7139', '#0078D4', '#808080']

# Explode the largest slice
explode = (0.05, 0, 0, 0, 0)  # Slightly separate Chrome

# Create figure
fig, ax = plt.subplots(figsize=(8, 6))

# Create pie with styling
wedges, texts, autotexts = ax.pie(
    market_share,
    labels=browsers,
    autopct='%1.1f%%',
    explode=explode,
    colors=colors,
    shadow=True,
    startangle=90  # Start from top
)

# Style the percentage text
for autotext in autotexts:
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')

ax.set_title("ğŸŒ Browser Market Share 2024", fontsize=14, fontweight='bold')

# Equal aspect ratio ensures circular pie
ax.axis('equal')

plt.show()

print("\nğŸ¨ Styling Features Used:")
print("  â€¢ Custom colors matching brand identities")
print("  â€¢ explode=(0.05,...) separates Chrome slice")
print("  â€¢ shadow=True adds depth")
print("  â€¢ startangle=90 starts from top")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ¨ PYTHONIC: Donut chart with center text (modern alternative)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ… Best: Donut chart is more modern and cleaner than traditional pie

import matplotlib.pyplot as plt

# Monthly expense breakdown
categories = ['Rent/Mortgage', 'Food', 'Transportation', 'Utilities', 'Entertainment', 'Savings']
amounts = [1500, 600, 300, 200, 250, 400]
colors = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12', '#9b59b6', '#1abc9c']

# Calculate total for center text
total = sum(amounts)

# Create figure
fig, ax = plt.subplots(figsize=(10, 7))

# Create donut chart (pie with center circle)
wedges, texts, autotexts = ax.pie(
    amounts,
    labels=categories,
    autopct=lambda pct: f'${int(pct/100 * total):,}',  # Show actual amounts
    colors=colors,
    startangle=90,
    pctdistance=0.75,  # Position of percentage labels
    wedgeprops=dict(width=0.5, edgecolor='white', linewidth=2)  # Creates donut
)

# Style text
for text in texts:
    text.set_fontsize(10)
for autotext in autotexts:
    autotext.set_fontsize(9)
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# Add center circle with total
center_circle = plt.Circle((0, 0), 0.35, fc='white')
ax.add_patch(center_circle)

# Add center text
ax.text(0, 0.05, 'Total', ha='center', va='center', fontsize=12, color='gray')
ax.text(0, -0.1, f'${total:,}', ha='center', va='center', fontsize=20, fontweight='bold')

ax.set_title("ğŸ’° Monthly Budget Breakdown", fontsize=14, fontweight='bold', pad=20)
ax.axis('equal')

plt.tight_layout()
plt.show()

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š When to Use Pie vs Bar Charts
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
print("\nğŸ¤” Pie Charts: Best Practices")
print("=" * 50)
print("âœ… USE pie charts when:")
print("  â€¢ Showing parts of a whole (percentages)")
print("  â€¢ You have 5 or fewer categories")
print("  â€¢ Values are significantly different")
print()
print("âŒ AVOID pie charts when:")
print("  â€¢ Comparing values precisely")
print("  â€¢ You have many categories")
print("  â€¢ Values are similar in size")
print()
print("ğŸ’¡ Pro tip: Bar charts are almost always clearer!")
````

## ğŸ“Š Section 6: Histograms

**Histograms** show the distribution of a single variable. Unlike bar charts, they show ranges (bins) of continuous data.

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸš« BAD PRACTICE: Histogram confusion
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âŒ Problem: Using wrong number of bins, no labels

import matplotlib.pyplot as plt
import random

# Generate random test scores
random.seed(42)
scores = [random.gauss(75, 10) for _ in range(100)]

# Too few bins - loses detail
plt.hist(scores, bins=3)
plt.show()

# âš ï¸ What's wrong here:
# - bins=3 is way too few (hides distribution shape)
# - No axis labels or title
# - Can't see the actual distribution pattern
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸŒ± NOVICE: Basic histogram with proper bins
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ… Better: Reasonable bins, labels, title

import matplotlib.pyplot as plt
import random

# Generate test scores (normal distribution around 75)
random.seed(42)
scores = [random.gauss(75, 10) for _ in range(100)]

# Create histogram with good number of bins
plt.hist(scores, bins=10)

# Add labels
plt.xlabel("Test Score")
plt.ylabel("Number of Students")
plt.title("Distribution of Test Scores")

plt.show()

print("\nâœ… Now we can see:")
print("  â€¢ Scores cluster around 75 (the mean)")
print("  â€¢ Bell-curve shape (normal distribution)")
print("  â€¢ Range is roughly 50-100")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”§ INTERMEDIATE: Styled histogram with statistics overlay
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ… Better: Colors, edge colors, and statistical annotations

import matplotlib.pyplot as plt
import random
import statistics

# Generate test scores
random.seed(42)
scores = [random.gauss(75, 10) for _ in range(100)]

# Calculate statistics
mean_score = statistics.mean(scores)
median_score = statistics.median(scores)
std_dev = statistics.stdev(scores)

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Create styled histogram
n, bins, patches = ax.hist(
    scores, 
    bins=15,
    color='steelblue',
    edgecolor='white',
    linewidth=1.2,
    alpha=0.7
)

# Add mean line
ax.axvline(mean_score, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_score:.1f}')
ax.axvline(median_score, color='green', linestyle=':', linewidth=2, label=f'Median: {median_score:.1f}')

# Labels and title
ax.set_xlabel("Test Score", fontsize=12)
ax.set_ylabel("Frequency", fontsize=12)
ax.set_title("ğŸ“Š Distribution of Test Scores with Statistics", fontsize=14, fontweight='bold')

# Add legend
ax.legend(loc='upper right')

# Add statistics text box
stats_text = f'n = {len(scores)}\nÏƒ = {std_dev:.1f}'
ax.text(0.02, 0.98, stats_text, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()

print("\nğŸ“ˆ Statistics Shown:")
print(f"  â€¢ Mean (average): {mean_score:.1f}")
print(f"  â€¢ Median (middle value): {median_score:.1f}")
print(f"  â€¢ Standard deviation: {std_dev:.1f}")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ¨ PYTHONIC: Multiple histograms with comparison
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ… Best: Compare distributions, density plot, professional styling

import matplotlib.pyplot as plt
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate data for two classes
class_a_scores = np.random.normal(75, 10, 150)  # Mean 75, std 10
class_b_scores = np.random.normal(82, 8, 150)   # Mean 82, std 8 (better class!)

# Create figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Left plot: Overlapping histograms
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
ax1 = axes[0]

# Plot both distributions
ax1.hist(class_a_scores, bins=20, alpha=0.5, color='#e74c3c', 
         edgecolor='darkred', label='Class A', density=True)
ax1.hist(class_b_scores, bins=20, alpha=0.5, color='#3498db', 
         edgecolor='darkblue', label='Class B', density=True)

# Add mean lines
ax1.axvline(np.mean(class_a_scores), color='#e74c3c', linestyle='--', linewidth=2)
ax1.axvline(np.mean(class_b_scores), color='#3498db', linestyle='--', linewidth=2)

ax1.set_xlabel("Test Score", fontsize=11)
ax1.set_ylabel("Density", fontsize=11)
ax1.set_title("ğŸ“Š Overlapping Comparison", fontsize=12, fontweight='bold')
ax1.legend(loc='upper left')
ax1.grid(True, alpha=0.3)

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Right plot: Side-by-side (step) histograms
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
ax2 = axes[1]

# Step histogram (outline only)
ax2.hist(class_a_scores, bins=20, histtype='step', linewidth=2.5, 
         color='#e74c3c', label=f'Class A (Î¼={np.mean(class_a_scores):.1f})')
ax2.hist(class_b_scores, bins=20, histtype='step', linewidth=2.5, 
         color='#3498db', label=f'Class B (Î¼={np.mean(class_b_scores):.1f})')

ax2.set_xlabel("Test Score", fontsize=11)
ax2.set_ylabel("Frequency", fontsize=11)
ax2.set_title("ğŸ“ˆ Step Histogram Comparison", fontsize=12, fontweight='bold')
ax2.legend(loc='upper left')
ax2.grid(True, alpha=0.3)

plt.suptitle("ğŸ“ Comparing Test Score Distributions Between Classes", 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š Comparison Statistics
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
print("\nğŸ“ˆ Distribution Comparison:")
print("=" * 50)
print(f"{'Metric':<20} {'Class A':<15} {'Class B':<15}")
print("-" * 50)
print(f"{'Mean':<20} {np.mean(class_a_scores):>14.1f} {np.mean(class_b_scores):>14.1f}")
print(f"{'Median':<20} {np.median(class_a_scores):>14.1f} {np.median(class_b_scores):>14.1f}")
print(f"{'Std Dev':<20} {np.std(class_a_scores):>14.1f} {np.std(class_b_scores):>14.1f}")
print(f"{'Min':<20} {np.min(class_a_scores):>14.1f} {np.min(class_b_scores):>14.1f}")
print(f"{'Max':<20} {np.max(class_a_scores):>14.1f} {np.max(class_b_scores):>14.1f}")

print("\nğŸ” Key Insight: Class B has higher scores AND lower variance!")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ… SOLUTION: Challenge 1 - Horizontal Bar Chart
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import matplotlib.pyplot as plt

languages = ['Python', 'JavaScript', 'Java', 'C++', 'Go', 'Rust']
popularity = [100, 95, 70, 55, 40, 35]
colors = ['#3776ab', '#f7df1e', '#f89820', '#00599C', '#00ADD8', '#dea584']

fig, ax = plt.subplots(figsize=(10, 6))

# Create horizontal bar chart
bars = ax.barh(languages, popularity, color=colors, edgecolor='white', linewidth=1.5)

# Add value labels
for bar, val in zip(bars, popularity):
    ax.text(val + 2, bar.get_y() + bar.get_height()/2, 
            f'{val}', va='center', fontsize=11, fontweight='bold')

ax.set_xlabel("Popularity Score", fontsize=12)
ax.set_title("ğŸ’» Programming Language Popularity 2024", fontsize=14, fontweight='bold')
ax.set_xlim(0, 115)
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# âœ… SOLUTION: Challenge 2 - Multi-Line Chart
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
company_a = [100, 105, 102, 110, 115, 120]
company_b = [80, 82, 90, 88, 95, 105]

fig, ax = plt.subplots(figsize=(10, 6))

# Plot both companies
ax.plot(months, company_a, color='#2ecc71', marker='o', linewidth=2.5, 
        markersize=8, label='Company A')
ax.plot(months, company_b, color='#e74c3c', marker='s', linewidth=2.5, 
        markersize=8, label='Company B')

# Styling
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Stock Price ($)", fontsize=12)
ax.set_title("ğŸ“ˆ Stock Price Comparison: H1 2024", fontsize=14, fontweight='bold')
ax.legend(loc='upper left')
ax.grid(True, linestyle='--', alpha=0.5)

# Add annotation for significant event
ax.annotate('Company B surge!', xy=('Mar', 90), xytext=('Apr', 82),
            arrowprops=dict(arrowstyle='->', color='#e74c3c'),
            fontsize=10, color='#e74c3c')

plt.tight_layout()
plt.show()
````

### What We Learned Today:

| Chart Type | Best For | Key Function |
|------------|----------|--------------|
| **Bar Chart** | Comparing categories | `plt.bar()` / `plt.barh()` |
| **Line Chart** | Showing trends over time | `plt.plot()` |
| **Scatter Plot** | Showing relationships | `plt.scatter()` |

| **Histogram** | Distribution of values | `plt.hist()` |

### Key Matplotlib Concepts:

```python
# The anatomy of a good chart:
fig, ax = plt.subplots(figsize=(10, 6))  # Create figure

ax.plot(x, y)                             # Plot data
ax.set_xlabel("X Axis Label")             # Label x-axis
ax.set_ylabel("Y Axis Label")             # Label y-axis
ax.set_title("Descriptive Title")         # Add title
ax.legend()                               # Add legend
ax.grid(True)                             # Add grid

plt.tight_layout()                        # Prevent overlap
plt.show()                                # Display
```
