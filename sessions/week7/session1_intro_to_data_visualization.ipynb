### Bar Chart (Optional Matplotlib)

```python
data = {'A':3,'B':7,'C':5}try:    import matplotlib.pyplot as plt    plt.bar(list(data.keys()), list(data.values()))    plt.title('Example Bar Chart')    plt.show()except Exception as e:    for k,v in data.items():        print(k + ' ' + ('#'*v))
```
