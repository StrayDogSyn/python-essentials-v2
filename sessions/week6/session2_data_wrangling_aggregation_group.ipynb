## ğŸ“Š The Challenge Dataset: Retail Sales

Your team has been given a month of retail sales data. Your mission: **Extract valuable business insights!**

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š CREATE THE RETAIL SALES DATASET
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

np.random.seed(42)  # For reproducibility

# Generate realistic sales data
n_records = 100
sales_data = pd.DataFrame({
    'transaction_id': range(10001, 10001 + n_records),
    'date': pd.date_range('2024-01-01', periods=n_records, freq='8H'),
    'store': np.random.choice(['Downtown', 'Mall', 'Airport', 'Suburb'], n_records),
    'department': np.random.choice(['Electronics', 'Clothing', 'Home', 'Food', 'Sports'], n_records),
    'product': np.random.choice(['Widget', 'Gadget', 'Gizmo', 'Thing', 'Item'], n_records),
    'quantity': np.random.randint(1, 10, n_records),
    'unit_price': np.random.choice([9.99, 19.99, 29.99, 49.99, 99.99, 199.99], n_records),
    'customer_type': np.random.choice(['Regular', 'Member', 'VIP'], n_records, p=[0.5, 0.35, 0.15]),
    'salesperson': np.random.choice(['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'], n_records)
})

# Calculate totals
sales_data['total'] = sales_data['quantity'] * sales_data['unit_price']
sales_data['day_of_week'] = sales_data['date'].dt.day_name()

print("ğŸ“Š RETAIL SALES DATASET")
print("=" * 70)
print(sales_data.head(10))
print(f"\nğŸ“ Shape: {sales_data.shape}")
print(f"ğŸ’° Total Revenue: ${sales_data['total'].sum():,.2f}")
```

## ğŸ Activity 1: Sales Analysis Race (15 min)

**Goal:** Answer business questions as quickly as possible!

### ğŸ“‹ Questions to Answer

Your team needs to find:
1. Total revenue by store
2. Average transaction value by department  
3. Top 3 salespeople by total revenue
4. Revenue by customer type
5. Busiest day of the week

**ğŸ­ Roles:**
- **Analyst** decides which question to tackle first
- **Engineer** writes the code
- **Validator** checks the results
- **Presenter** documents findings

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ ACTIVITY 1: ANSWER THE BUSINESS QUESTIONS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

df = sales_data.copy()

print("ğŸ SALES ANALYSIS RESULTS")
print("=" * 60)

# Q1: Total revenue by store
q1 = df.groupby('store')['total'].sum().sort_values(ascending=False)
print("\n1ï¸âƒ£ Revenue by Store:")
print(q1)

# Q2: Average transaction by department
q2 = df.groupby('department')['total'].mean().round(2)
print("\n2ï¸âƒ£ Average Transaction by Department:")
print(q2)

# Q3: Top 3 salespeople
q3 = df.groupby('salesperson')['total'].sum().nlargest(3)
print("\n3ï¸âƒ£ Top 3 Salespeople:")
print(q3)

# Q4: Revenue by customer type
q4 = df.groupby('customer_type')['total'].sum()
print("\n4ï¸âƒ£ Revenue by Customer Type:")
print(q4)

# Q5: Busiest day
q5 = df.groupby('day_of_week')['transaction_id'].count().sort_values(ascending=False)
print("\n5ï¸âƒ£ Transactions by Day of Week:")
print(q5)
print("=" * 60)
```

## ğŸ”§ Activity 2: Advanced GroupBy Challenge (15 min)

**Goal:** Use multi-level grouping and pivot tables!

### ğŸ“‹ Challenge Tasks

1. Create a pivot table: Store Ã— Department (total revenue)
2. Find the best salesperson in EACH store
3. Calculate the percentage of revenue from VIP customers per store
4. Find departments where average transaction > $100

**ğŸ­ Rotate roles from Activity 1!**

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”§ ACTIVITY 2: ADVANCED GROUPBY CHALLENGES
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

df = sales_data.copy()

print("ğŸ”§ ADVANCED GROUPBY RESULTS")
print("=" * 60)

# Challenge 1: Pivot table - Store Ã— Department
pivot = df.pivot_table(
    values='total',
    index='store',
    columns='department',
    aggfunc='sum',
    fill_value=0
)
print("\n1ï¸âƒ£ Revenue Pivot: Store Ã— Department")
print(pivot.round(2))

# Challenge 2: Best salesperson per store
best_per_store = df.groupby(['store', 'salesperson'])['total'].sum().reset_index()
idx = best_per_store.groupby('store')['total'].idxmax()
print("\n2ï¸âƒ£ Best Salesperson per Store:")
print(best_per_store.loc[idx, ['store', 'salesperson', 'total']])

# Challenge 3: VIP percentage per store
store_totals = df.groupby('store')['total'].sum()
vip_totals = df[df['customer_type'] == 'VIP'].groupby('store')['total'].sum()
vip_pct = (vip_totals / store_totals * 100).round(2)
print("\n3ï¸âƒ£ VIP Revenue Percentage by Store:")
print(vip_pct)

# Challenge 4: High-value departments
dept_avg = df.groupby('department')['total'].mean()
high_value_depts = dept_avg[dept_avg > 100]
print("\n4ï¸âƒ£ Departments with Avg Transaction > $100:")
print(high_value_depts.round(2))
print("=" * 60)
```

## ğŸ† Activity 3: Build a Sales Report (15 min)

**Goal:** Create a comprehensive sales report function!

### ğŸ“‹ Report Requirements

Build a function that generates a complete sales summary including:
- Total revenue and transaction count
- Top performing store
- Top performing department  
- Best salesperson
- Customer type breakdown

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ† ACTIVITY 3: BUILD A SALES REPORT FUNCTION
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def generate_sales_report(df):
    """
    Generate a comprehensive sales report from transaction data.
    
    Args:
        df: DataFrame with sales transactions
        
    Returns:
        Dictionary with report metrics
    """
    report = {}
    
    # Overall metrics
    report['total_revenue'] = df['total'].sum()
    report['transaction_count'] = len(df)
    report['avg_transaction'] = df['total'].mean()
    
    # Top performers
    store_sales = df.groupby('store')['total'].sum()
    report['top_store'] = store_sales.idxmax()
    report['top_store_revenue'] = store_sales.max()
    
    dept_sales = df.groupby('department')['total'].sum()
    report['top_department'] = dept_sales.idxmax()
    
    person_sales = df.groupby('salesperson')['total'].sum()
    report['top_salesperson'] = person_sales.idxmax()
    report['top_salesperson_revenue'] = person_sales.max()
    
    # Customer breakdown
    customer_pct = df.groupby('customer_type')['total'].sum() / df['total'].sum() * 100
    report['customer_breakdown'] = customer_pct.round(2).to_dict()
    
    return report


# Generate and display report
print("ğŸ† SALES REPORT")
print("=" * 60)

report = generate_sales_report(sales_data)

print(f"\nğŸ’° OVERALL METRICS")
print(f"   Total Revenue: ${report['total_revenue']:,.2f}")
print(f"   Transactions: {report['transaction_count']}")
print(f"   Avg Transaction: ${report['avg_transaction']:.2f}")

print(f"\nğŸ† TOP PERFORMERS")
print(f"   Best Store: {report['top_store']} (${report['top_store_revenue']:,.2f})")
print(f"   Best Department: {report['top_department']}")
print(f"   Best Salesperson: {report['top_salesperson']} (${report['top_salesperson_revenue']:,.2f})")

print(f"\nğŸ‘¥ CUSTOMER BREAKDOWN")
for ctype, pct in report['customer_breakdown'].items():
    print(f"   {ctype}: {pct}%")
    
print("=" * 60)
```
