# ğŸ“ˆ Activity 1: Advanced Multi-Series Plotting

## Concept Overview

Multi-series charts compare **multiple data sets** on a single figure. Advanced features include:
- Custom markers and line styles per series
- Finding and annotating the **global maximum**
- Professional legends and styling

### ğŸ¯ Goal
Create a multi-series chart that **identifies and annotates the maximum value** across all series.

---

## ğŸš« Bad Practice: Minimal Multi-Series

Issues with this approach:
- No axis labels
- No clear distinction between series styles
- No identification of key points

````python
# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘              ğŸš« BAD PRACTICE: Minimal Multi-Series                          â•‘
# â•‘              Issues: No labels, no styling, no key point identification     â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Sample data - three time series
series = {'A': [1, 2, 3, 5, 4], 'B': [2, 3, 1, 2, 6], 'C': [3, 1, 2, 4, 3]}

# âŒ Problems with this approach:
# - No axis labels - what do these numbers represent?
# - No clear visual distinction between series
# - Missing identification of the maximum value
# - Minimal legend with no positioning

try:
    import matplotlib.pyplot as plt
    for name, values in series.items():
        plt.plot(values, label=name)
    plt.legend()
    plt.title('Three Series')
    plt.show()
except Exception:
    for name, values in series.items():
        print(name, values)
````

## ğŸŒ± Novice: Adding Basic Styling and Labels

````python
# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘              ğŸŒ± NOVICE: Adding Basic Styling and Labels                     â•‘
# â•‘              Improvement: Axis labels, markers, and better legend           â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

series = {'A': [1, 2, 3, 5, 4], 'B': [2, 3, 1, 2, 6], 'C': [3, 1, 2, 4, 3]}
x_values = [0, 1, 2, 3, 4]  # Time points

try:
    import matplotlib.pyplot as plt
    
    # Create figure with size
    plt.figure(figsize=(10, 6))
    
    # âœ… Better: Different markers for each series
    markers = ['o', 's', '^']
    for i, (name, values) in enumerate(series.items()):
        plt.plot(x_values, values, label=f'Series {name}', marker=markers[i])
    
    # âœ… Better: Add axis labels
    plt.xlabel('Time Period')
    plt.ylabel('Value')
    plt.title('Three Series Comparison')
    
    # âœ… Better: Position the legend
    plt.legend(loc='upper left')
    plt.grid(True, alpha=0.3)
    plt.show()
    
except ImportError:
    print("ğŸ“ˆ Series Comparison (Text Mode)")
    print("=" * 40)
    for name, values in series.items():
        print(f"Series {name}: {values}")
````

## ğŸ”§ Intermediate: Finding and Annotating the Global Maximum

````python
# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘              ğŸ”§ INTERMEDIATE: Annotating Global Maximum                     â•‘
# â•‘              New: Find max across ALL series and annotate it                â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

series = {'A': [1, 2, 3, 5, 4], 'B': [2, 3, 1, 2, 6], 'C': [3, 1, 2, 4, 3]}
x_values = [0, 1, 2, 3, 4]

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Step 1: Find the global maximum across all series
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
global_max = float('-inf')
max_series = None
max_x_idx = None

for name, values in series.items():
    for idx, val in enumerate(values):
        if val > global_max:
            global_max = val
            max_series = name
            max_x_idx = idx

print(f"ğŸ” Global Maximum Found:")
print(f"   Value: {global_max}")
print(f"   Series: {max_series}")
print(f"   Position: x={max_x_idx}")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Step 2: Plot with annotation
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
try:
    import matplotlib.pyplot as plt
    
    plt.figure(figsize=(10, 6))
    
    colors = {'A': 'steelblue', 'B': 'coral', 'C': 'seagreen'}
    markers = {'A': 'o', 'B': 's', 'C': '^'}
    
    for name, values in series.items():
        plt.plot(x_values, values, label=f'Series {name}', 
                 marker=markers[name], color=colors[name], 
                 linewidth=2, markersize=8)
    
    # âœ… Key Feature: Annotate the global maximum
    plt.annotate(f'MAX: {global_max}\n(Series {max_series})',
                 xy=(max_x_idx, global_max),
                 xytext=(max_x_idx + 0.5, global_max + 0.5),
                 fontsize=10,
                 color='darkred',
                 fontweight='bold',
                 arrowprops=dict(arrowstyle='->', color='red', lw=2))
    
    plt.xlabel('Time Period', fontsize=12)
    plt.ylabel('Value', fontsize=12)
    plt.title('Multi-Series with Global Maximum Annotated', fontsize=14, fontweight='bold')
    plt.legend(loc='upper left')
    plt.grid(True, alpha=0.3)
    plt.show()
    
except ImportError:
    print("\nğŸ“Š Text Visualization:")
    print("=" * 50)
    for name, values in series.items():
        row = f"Series {name}: "
        for i, v in enumerate(values):
            marker = "â˜…" if (name == max_series and i == max_x_idx) else "â€¢"
            row += f"{v}{marker} "
        print(row)
    print(f"\nâ˜… = Global Maximum ({global_max} in Series {max_series})")
````

## âœ¨ Pythonic: Professional Multi-Series with Comprehensive Features

````python
# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘              âœ¨ PYTHONIC: Professional Multi-Series Chart                   â•‘
# â•‘              Features: Reusable function, comprehensive annotations,        â•‘
# â•‘                        graceful fallbacks, statistics display               â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def create_multi_series_chart(series_data, x_labels=None, title="Multi-Series Chart"):
    """
    Create a professional multi-series line chart with global max annotation.
    
    Args:
        series_data: Dict of {series_name: [values]}
        x_labels: Optional list of x-axis labels
        title: Chart title
    
    Returns:
        Dict with chart statistics
    """
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # Calculate statistics for all series
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    stats = {'series': {}, 'global_max': {'value': float('-inf')}}
    
    for name, values in series_data.items():
        series_max = max(values)
        series_min = min(values)
        series_avg = sum(values) / len(values)
        max_idx = values.index(series_max)
        
        stats['series'][name] = {
            'max': series_max,
            'min': series_min,
            'avg': series_avg,
            'max_idx': max_idx
        }
        
        if series_max > stats['global_max']['value']:
            stats['global_max'] = {
                'value': series_max,
                'series': name,
                'idx': max_idx
            }
    
    n_points = len(next(iter(series_data.values())))
    x_vals = list(range(n_points))
    
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # Visualization (matplotlib or text fallback)
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    try:
        import matplotlib.pyplot as plt
        
        fig, ax = plt.subplots(figsize=(12, 7))
        
        # Style configuration
        colors = ['steelblue', 'coral', 'seagreen', 'orchid', 'goldenrod']
        markers = ['o', 's', '^', 'D', 'v']
        
        for i, (name, values) in enumerate(series_data.items()):
            color = colors[i % len(colors)]
            marker = markers[i % len(markers)]
            
            ax.plot(x_vals, values, 
                    label=f'{name} (avg: {stats["series"][name]["avg"]:.1f})',
                    marker=marker, color=color, linewidth=2.5, 
                    markersize=10, markeredgecolor='white', markeredgewidth=1)
        
        # Annotate global maximum with enhanced styling
        gmax = stats['global_max']
        ax.annotate(f'GLOBAL MAX\n{gmax["value"]}\n(Series {gmax["series"]})',
                    xy=(gmax['idx'], gmax['value']),
                    xytext=(gmax['idx'] + 0.8, gmax['value'] + 0.8),
                    fontsize=11,
                    color='white',
                    fontweight='bold',
                    ha='center',
                    bbox=dict(boxstyle='round,pad=0.5', facecolor='crimson', alpha=0.9),
                    arrowprops=dict(arrowstyle='->', color='crimson', lw=2.5))
        
        # Highlight the max point
        ax.scatter([gmax['idx']], [gmax['value']], 
                   c='crimson', s=200, zorder=5, marker='*', edgecolors='gold', linewidths=2)
        
        # Labels and styling
        ax.set_xlabel('Time Period', fontsize=12, fontweight='bold')
        ax.set_ylabel('Value', fontsize=12, fontweight='bold')
        ax.set_title(title, fontsize=16, fontweight='bold', pad=15)
        
        if x_labels:
            ax.set_xticks(x_vals)
            ax.set_xticklabels(x_labels)
        
        ax.legend(loc='upper left', fontsize=10, framealpha=0.9)
        ax.grid(True, alpha=0.3, linestyle='--')
        ax.set_facecolor('#f8f9fa')
        
        plt.tight_layout()
        plt.show()
        
    except ImportError:
        # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
        # Text-based fallback visualization
        # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
        print(f"\n{'â•' * 60}")
        print(f"ğŸ“Š {title} (Text Mode)")
        print(f"{'â•' * 60}")
        
        gmax = stats['global_max']
        
        for name, values in series_data.items():
            s = stats['series'][name]
            print(f"\nğŸ“ˆ Series {name}:")
            print(f"   Values: ", end="")
            for i, v in enumerate(values):
                if name == gmax['series'] and i == gmax['idx']:
                    print(f"[â˜…{v}â˜…]", end=" ")
                else:
                    print(f"{v}", end=" ")
            print(f"\n   Stats: Min={s['min']} | Avg={s['avg']:.1f} | Max={s['max']}")
        
        print(f"\n{'â”€' * 60}")
        print(f"ğŸ† GLOBAL MAXIMUM: {gmax['value']} (Series {gmax['series']}, Position {gmax['idx']})")
        print(f"{'â•' * 60}")
    
    return stats

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Demo: Use the reusable function
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
series = {'A': [1, 2, 3, 5, 4], 'B': [2, 3, 1, 2, 6], 'C': [3, 1, 2, 4, 3]}
x_labels = ['Week 1', 'Week 2', 'Week 3', 'Week 4', 'Week 5']

result = create_multi_series_chart(series, x_labels, "Weekly Performance Comparison")

print("\nğŸ“‹ Returned Statistics:")
print(f"   Global Max: {result['global_max']}")
````

## ğŸ”§ Intermediate: Computing and Annotating Category Centers

````python
# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘              ğŸ”§ INTERMEDIATE: Annotating Category Centers                   â•‘
# â•‘              New: Calculate average (center) of each category               â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

points = [(1, 2, 'A'), (2, 2, 'B'), (3, 4, 'A'), (4, 3, 'B'),
          (1.5, 3, 'A'), (3.5, 2.5, 'B'), (2, 3.5, 'A'), (4.5, 4, 'B')]

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Step 1: Organize and calculate centers
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
categories = {}
for x, y, cat in points:
    if cat not in categories:
        categories[cat] = {'x': [], 'y': []}
    categories[cat]['x'].append(x)
    categories[cat]['y'].append(y)

# Calculate centers (averages)
centers = {}
for cat, data in categories.items():
    center_x = sum(data['x']) / len(data['x'])
    center_y = sum(data['y']) / len(data['y'])
    centers[cat] = (center_x, center_y)
    print(f"ğŸ“ Category {cat} center: ({center_x:.2f}, {center_y:.2f})")

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Step 2: Visualize with center annotations
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
try:
    import matplotlib.pyplot as plt
    
    fig, ax = plt.subplots(figsize=(10, 7))
    
    colors = {'A': 'steelblue', 'B': 'coral'}
    
    for cat, data in categories.items():
        # Plot scatter points
        ax.scatter(data['x'], data['y'], 
                   label=f'Category {cat}',
                   c=colors.get(cat, 'gray'),
                   s=120, alpha=0.6, edgecolors='white', linewidth=1)
        
        # Plot center as larger marker
        cx, cy = centers[cat]
        ax.scatter([cx], [cy], c=colors.get(cat, 'gray'), 
                   s=300, marker='X', edgecolors='black', linewidth=2, zorder=5)
        
        # Annotate the center
        ax.annotate(f'Center {cat}\n({cx:.1f}, {cy:.1f})',
                    xy=(cx, cy),
                    xytext=(cx + 0.5, cy + 0.5),
                    fontsize=10,
                    fontweight='bold',
                    arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
    
    ax.set_xlabel('X Value', fontsize=12)
    ax.set_ylabel('Y Value', fontsize=12)
    ax.set_title('Scatter Plot with Category Centers', fontsize=14, fontweight='bold')
    ax.legend(loc='upper left')
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
except ImportError:
    print("\nğŸ“ˆ Category Analysis (Text Mode)")
    print("=" * 45)
    for cat, data in categories.items():
        cx, cy = centers[cat]
        print(f"\nCategory {cat}:")
        print(f"   Points: {list(zip(data['x'], data['y']))}")
        print(f"   Center: ({cx:.2f}, {cy:.2f})")
````

## âœ¨ Pythonic: Professional Categorized Scatter with Statistics

````python
# â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
# â•‘              âœ¨ PYTHONIC: Professional Categorized Scatter                  â•‘
# â•‘              Features: Reusable function, comprehensive stats,              â•‘
# â•‘                        distance calculation, graceful fallbacks             â•‘
# â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import math

def create_categorized_scatter(points, title="Categorized Scatter Plot"):
    """
    Create a professional categorized scatter plot with center annotations.
    
    Args:
        points: List of (x, y, category) tuples
        title: Chart title
    
    Returns:
        Dict with category statistics including centers and spread
    """
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # Organize data and calculate comprehensive statistics
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    categories = {}
    for x, y, cat in points:
        if cat not in categories:
            categories[cat] = {'x': [], 'y': []}
        categories[cat]['x'].append(x)
        categories[cat]['y'].append(y)
    
    stats = {}
    for cat, data in categories.items():
        n = len(data['x'])
        center_x = sum(data['x']) / n
        center_y = sum(data['y']) / n
        
        # Calculate spread (average distance from center)
        distances = [math.sqrt((x - center_x)**2 + (y - center_y)**2) 
                     for x, y in zip(data['x'], data['y'])]
        avg_spread = sum(distances) / n if n > 0 else 0
        
        stats[cat] = {
            'center': (center_x, center_y),
            'count': n,
            'spread': avg_spread,
            'x_range': (min(data['x']), max(data['x'])),
            'y_range': (min(data['y']), max(data['y']))
        }
    
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # Visualization
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    try:
        import matplotlib.pyplot as plt
        from matplotlib.patches import Circle
        
        fig, ax = plt.subplots(figsize=(11, 8))
        
        colors = ['steelblue', 'coral', 'seagreen', 'orchid', 'goldenrod']
        
        for i, (cat, data) in enumerate(categories.items()):
            color = colors[i % len(colors)]
            s = stats[cat]
            
            # Plot scatter points
            ax.scatter(data['x'], data['y'],
                       label=f'{cat} (n={s["count"]})',
                       c=color, s=150, alpha=0.6,
                       edgecolors='white', linewidth=1.5)
            
            # Plot center
            cx, cy = s['center']
            ax.scatter([cx], [cy], c=color, s=350, marker='â˜…',
                       edgecolors='black', linewidth=2, zorder=5)
            
            # Draw spread circle (visual representation of average spread)
            circle = Circle((cx, cy), s['spread'], fill=False,
                            color=color, linestyle='--', linewidth=2, alpha=0.5)
            ax.add_patch(circle)
            
            # Annotate center with statistics
            ax.annotate(f'{cat} Center\n({cx:.2f}, {cy:.2f})\nSpread: {s["spread"]:.2f}',
                        xy=(cx, cy),
                        xytext=(cx + s['spread'] + 0.3, cy + 0.3),
                        fontsize=9,
                        fontweight='bold',
                        color='white',
                        bbox=dict(boxstyle='round,pad=0.4', facecolor=color, alpha=0.9),
                        arrowprops=dict(arrowstyle='->', color=color, lw=2))
        
        ax.set_xlabel('X Value', fontsize=12, fontweight='bold')
        ax.set_ylabel('Y Value', fontsize=12, fontweight='bold')
        ax.set_title(title, fontsize=15, fontweight='bold', pad=15)
        ax.legend(loc='upper right', fontsize=10)
        ax.grid(True, alpha=0.3, linestyle='--')
        ax.set_facecolor('#f8f9fa')
        ax.set_aspect('equal', adjustable='box')
        
        plt.tight_layout()
        plt.show()
        
    except ImportError:
        # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
        # Text fallback with comprehensive statistics
        # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
        print(f"\n{'â•' * 60}")
        print(f"ğŸ“Š {title} (Text Mode)")
        print(f"{'â•' * 60}")
        
        for cat, s in stats.items():
            cx, cy = s['center']
            print(f"\nğŸ”µ Category {cat}:")
            print(f"   Points: {s['count']}")
            print(f"   Center: ({cx:.2f}, {cy:.2f})")
            print(f"   Spread: {s['spread']:.2f}")
            print(f"   X Range: {s['x_range'][0]:.1f} to {s['x_range'][1]:.1f}")
            print(f"   Y Range: {s['y_range'][0]:.1f} to {s['y_range'][1]:.1f}")
        
        # Visual representation
        print(f"\n{'â”€' * 60}")
        print("ğŸ“ Category Centers Comparison:")
        for cat, s in stats.items():
            cx, cy = s['center']
            bar = "â–ˆ" * int(s['spread'] * 5)
            print(f"   {cat}: ({cx:.1f}, {cy:.1f}) Spread: {bar} {s['spread']:.2f}")
        
        print(f"{'â•' * 60}")
    
    return {'categories': categories, 'stats': stats}

# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Demo: Use the reusable function
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
points = [(1, 2, 'A'), (2, 2, 'B'), (3, 4, 'A'), (4, 3, 'B'),
          (1.5, 3, 'A'), (3.5, 2.5, 'B'), (2, 3.5, 'A'), (4.5, 4, 'B')]

result = create_categorized_scatter(points, "Category Analysis with Center & Spread")

print("\nğŸ“‹ Returned Statistics:")
for cat, s in result['stats'].items():
    print(f"   Category {cat}: center={s['center']}, spread={s['spread']:.2f}")
````
